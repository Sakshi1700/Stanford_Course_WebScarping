```python
import selenium
from selenium import webdriver
import time
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
import pprint
import re
from selenium.webdriver.common.by import By
```

```python
l = {'https://bulletin.stanford.edu/programs/CS-MS'}
```

```python
url = []
```

```python
driver = selenium.webdriver.Chrome(r'C:\Users\Sakshi\Downloads\chromedriver_win32/chromedriver.exe')
```

```python
for link in l:
    driver.get(link)
    lii =driver.find_element(By.XPATH,'//*[@id="main-content"]/div/div[2]/article/div[2]/div/div/button[2]/span')
    #perform click
    lii.click()
    
    res1 = driver.page_source.encode('utf-8')
    soup = bs(res1,'html.parser')
    
    #getting the url
    a = soup.find_all('a',class_='custom-link')
    for j in a:
        url.append(j.get('href'))
    
```

```python
url
```

```python
result = list(map(lambda x : 'https://bulletin.stanford.edu' + x, url))
```

```python
del result[0:2]
```

```python
course = []
course_description = []
```

```python

        
   # co = soup.find_all('div',{'class':'field-value'})
    #for k in co:
     #   course_description.append(k.get('div'))
        

for i in result:
    driver.get(i)
    res1 = driver.page_source.encode('utf-8')
    soup = bs(res1,'html.parser')
    
    co = soup.find_all('h1',{'class':'pb-2 mb-0'})
    for k in co:
        course.append(k.text)
        
    des = soup.find_all('div',{'class':'field-value'})
    f_ele = des[0]
    for d in f_ele:
        course_description.append(d)
```

```python
import pandas as pd

dictt = {'Course' : course, 'Description' : course_description}
df2 = pd.DataFrame(dictt)

df2.to_csv('Courses_List.csv')
```
